```python
# After studying Numpy and Pandas for Data Analysis, now lets study about Matplotlib
# for data visualization.
# Matplotlib is the most popular plotting library for python.
# It gives you control over every aspect of a figure.
# Matplotlib works extremely well with pandas and numpy arrays.

# matplotlib.org/gallery.html
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
# to see the plots in jupyter notebook
# without explicitly typing plt.show()
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
# There are two ways to create plots using matplotlib
# 1. Functional Method
# 2. Object Oriented based
```

```python
# Lets first see functional method
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
# plt.show()
```

```python
# Creating multiplots on the same canvas
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

```python
# Object Oriented Way -> More better
# Main idea is to create figure object and just call methods of it.
fig = plt.figure() # imaginary canvas

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # [left, bottom, width, height] in relation to the empty canvas
axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
# adding 2 figures on same canvas
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) 
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y)
axes1.set_title('Larger Plot')

axes2.plot(y, x)
axes2.set_title('Smaller Plot')
```

```python
# Now We will see how to create subplots using object oriented approach
# which will give us a lot more control.
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2) # this will add_axes automatically for you

for current_ax in axes:
    current_ax.plot(x, y)
# plt.tight_layout() to manage overlapping of subplots
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y)
axes[0].set_title('First Plot')

axes[1].plot(y, x)
axes[1].set_title('Second Plot')

plt.tight_layout()
```

```python
# Figure Size and DPI
fig = plt.figure(figsize=(8, 2), dpi = 2000) # width, height in inches

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

```python
# can work with subplots also
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 2), dpi= 200)

axes[0].plot(x, y)
axes[1].plot(y, x)

plt.tight_layout()
```

```python
# saving a figure to a file

fig.savefig('my_picture.png')
```

```python
# legends -> labelled text

fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, label='X Squared')
ax.plot(x, x**3, label='X Cubed')

# ax.legend(loc='best') # will match label to the line
ax.legend(loc=(0.1, 0.1))
```
