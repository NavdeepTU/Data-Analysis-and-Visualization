```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.set_style('white')
sns.countplot(x='day', data=tips)
```

```python
sns.set_style('ticks')
sns.countplot(x='day', data=tips)
```

```python
sns.set_style('darkgrid')
sns.countplot(x='day', data=tips)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='day', data=tips)
```

```python
sns.set_style('ticks')
sns.countplot(x='day', data=tips)
sns.despine() # by default top and right spines will be removed
```

```python
sns.set_style('ticks')
sns.countplot(x='day', data=tips)
sns.despine(left=True, bottom=True) # all spines will be removed
```

```python
# Size ans aspect ratio
```

```python
plt.figure(figsize=(12,3)) # can use matplotlib functions to change figure size
sns.countplot(x='day', data=tips)
```

```python
# Scale and context of text
```

```python
sns.set_context('poster') # by default notebook
sns.countplot(x='day', data=tips)
```

```python
plt.figure(figsize=(10, 15))
sns.set_context('poster', font_scale=2)
sns.countplot(x='day', data=tips)
```

```python
# Pallets and colors
```

```python
sns.set_context('notebook')
sns.lmplot(x='total_bill', y='tip', data=tips, hue='smoker', palette='coolwarm')
# checkout matplotlib colormap for more palette/color strings
```
