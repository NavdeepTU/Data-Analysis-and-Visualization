```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
# Now we will learn ho to use seaborns grid capability to automate
# subplots based of featuers in out data.
```

```python
# PairGrid gives us the the capability to customize pairplots
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot) # distribution plot for diagnol
g.map_upper(plt.scatter) # scatter plot for upper half
g.map_lower(sns.kdeplot) # kde plot for lower half
```

```python
#FacetGrid
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(sns.distplot, 'total_bill')
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(plt.scatter, 'total_bill', 'tip') 
# if using a plot that needs x and y both then specify them as second and third argument
# of map function
```
