```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Linear Model plot
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='smoker')
```

```python
# changing marker styles
# since seaborn is built on top of matplotlib we can change the styling
# of the plot
sns.lmplot(x='total_bill', y='tip', data=tips, hue='smoker', markers=['o', 'v'],
          scatter_kws={'s':100})
```

```python
# seperating smoker using a column instead of color
# this is essentially doing grid calls and FacetGrid calls
# little more automatic
sns.lmplot(x='total_bill', y='tip', data=tips, col='smoker', row='time')
```

```python
# can also add hue to a column
sns.lmplot(x='total_bill', y='tip', data=tips, col='smoker', hue='time')
```

```python
# changing size and aspect ratio of seaborn plots
sns.lmplot(x='total_bill', y='tip', data=tips, col='smoker', hue='time',
          aspect=0.6, size=8)
```
