```python
# Seaborn is a statistical plotting library build on top of Matplotlib.
# It has beautiful default styles
# It also is designed to work very well with pandas dataframe objects.

# conda install seaborn 
# pip install seaborn

# Official documentation -> http://seaborn.pydata.org/api.html
```

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips') # Seaborn has buch of built in datasets
```

```python
tips.head()
```

```python
# Distribution plots
# It allows us to show the distribution of a univariate set of observations.

sns.distplot(tips['total_bill'], kde=False, bins=40)
```

```python
# Joint Plot
# Joint plots allows you to match up dist plots to bivariate data
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips) # default scatter
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

```python
# Pair Plots
# They are going to plot the pair wise relationships accross the DataFrame
# Pair Plot will basically plot joint plot for every 
# possible combination of numerical columns in the dataframe. 
```

```python
sns.pairplot(tips, hue='smoker', palette='coolwarm') # used quite often
# will touch on colors and palette a lot more in the end
```

```python
# Rug Plot
# It will show dashes based on number of points
# It is used to build kde plots
# We won't be used them much
sns.rugplot(tips['total_bill']) 
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill']) # If you dont want the bins
```
