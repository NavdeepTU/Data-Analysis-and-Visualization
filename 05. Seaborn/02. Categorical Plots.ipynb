```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Bar Plots
# x-axis -> some categorical data
# y-axis -> some numerical data
```

```python
sns.barplot(x='day', y='total_bill', data=tips) # by default estimator(y-axis) is mean
```

```python
import numpy as np
sns.barplot(x='smoker', y='total_bill', data=tips, estimator=np.std)
# we can add our own estimator
```

```python
# Count Plot
# It is same as bar plot but the estimator is 
# basically calculating the number of instances
```

```python
sns.countplot(x='time', data=tips) # visual representation of value_counts() function
```

```python
# Box Plots
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
# can add hue attribute to boxplot
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
# People on friday tend to pay more if they are not smokers vs 
# if they are smokers.
```

```python
# Violin Plot
# They are also box plots that include all the points(including outliers)
# and show percentile values along with kde plots for each
# category of data.
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker', split=True)
```

```python
# Strip Plot
# # Its just a scatter plot based off of a category
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='smoker', split=True)
# A plot like this could be harder to interpret vs a boxplot
# or a simple violin plot.
```

```python
# Swarm Plot
# They combine the idea of a strip plot and a violin plot
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
# Don't use swarm plots for very large datasets as computation gets expensive
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
```

```python
# Cat Plots
# Cat Plot is just a generally form of all the above plots
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='bar')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='violin')
```
