```python
# Most popular plotting library for Python
# It gives you control over every aspect of a figure
# Works very well with pandas and numpy arrays
# Official website : matplotlib.org -> gallery
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
y = x**2
```

```python
x
```

```python
y
```

```python
# FUNCTIONAL way of plotting
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
# plt.show()  no need of this statement if using jupyter notebook
```

```python
plt.subplot(1,2,1)
plt.plot(x, y, 'r')

plt.subplot(1,2,2)
plt.plot(y,x, 'b')
```

```python
# Object Oriented Method
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y)
axes1.set_title('LARGER_PLOT')

axes2.plot(y, x)
axes2.set_title('SMALLER PLOT')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2) #adds axes automatically for you

axes[0].plot(x, y)
axes[0].set_title('First Plot')

axes[1].plot(y, x)
axes[1].set_title('Second, Plot')

plt.tight_layout()
```

```python
#Figure size and DPI
fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2))

axes[0].plot(x,y)

axes[1].plot(y, x)

plt.tight_layout()
```

```python
# Saving a figure to a file
fig.savefig('my_picture.png', dpi=200) # Increase the DPI if you want more details
```

```python
# Labels and Legends
fig2 = plt.figure()

ax = fig2.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, label='X Squared')
ax.plot(x, x**3, label='X Cubed')

ax.legend(loc = 0) # Best Location
```
