```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0, 100)
y = x*2
z = x**2
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0, 0, 1, 1])
ax1.plot(x, y)

ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])
ax2.plot(x, y)
```

```python
fig1 = plt.figure()

ax1 = fig1.add_axes([0, 0, 1, 1])
ax1.plot(x, z, color='blue')
ax1.set_xlabel('X')
ax1.set_ylabel('Z')

ax2 = fig1.add_axes([0.2, 0.5, 0.4, 0.4])
ax2.plot(x, z, color='blue')
ax2.set_xlim([20, 22])
ax2.set_ylim([300, 600])
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('zoom')

# Second Figure
fig2 = plt.figure()

ax1 = fig2.add_axes([0, 0, 1, 1])
ax1.plot(x, z, color='blue')
ax1.set_xlabel('X')
ax1.set_ylabel('Z')

ax2 = fig2.add_axes([0.2, 0.5, 0.4, 0.4])
ax2.plot(x, z, color='blue')
ax2.set_xlim([20, 22])
ax2.set_ylim([300, 600])
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('zoom')
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2)

axs[0].plot(x, y, color='blue', lw=4, ls='--')
axs[1].plot(x, z, color='red', lw=4)
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(20, 10))

axs[0].plot(x, y, color='blue', lw=4, ls='--') 
axs[1].plot(x, z, color='red', lw=4)
```
