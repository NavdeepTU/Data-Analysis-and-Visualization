```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
df3 = pd.read_csv('df3.csv')
```

```python
df3.info()
```

```python
df3.head()
```

```python
# Create a scatter plot of b vs a
df3.plot.scatter(x='a', y='b', s=50, c='red', figsize=(12, 3))
```

```python
# Create a histogram of the 'a' column.
df3['a'].plot.hist()
```

```python
plt.style.use('ggplot')
df3['a'].plot.hist(bins=30, alpha=0.5)
```

```python
# Create a boxplot comparing the a and b columns.
df3[['a', 'b']].plot.box()
```

```python
# Create a kde plot of the 'd' column
df3['d'].plot.kde()
```

```python
# Figure out how to increase the linewidth and make the linestyle dashed. 
# (Note: You would usually not dash a kde plot line)

df3['d'].plot.kde(lw=3, ls='--', c='red')
```

```python
# Create an area plot of all the columns for just the rows up to 30. (hint: use .iloc).
df3.iloc[:30].plot.area(alpha=0.5)
```

```python
# Notice how the legend in our previous figure overlapped some of actual diagram. 
# Can you figure out how to display the legend outside of the plot
df3.iloc[:30].plot.area(alpha=0.4)
plt.legend(bbox_to_anchor=(1.0, 0.5))
```
